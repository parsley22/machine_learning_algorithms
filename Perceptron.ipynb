```python
import numpy as np
from sklearn.datasets import make_blobs, make_moons
import matplotlib.pyplot as plt

%load_ext nb_black
```

```python
X, y = make_blobs(n_samples=100, centers=2, cluster_std=2)
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
class perceptron:
    def __init__(self, lr=0.25):
        self.lr = lr

        self.weights = 0

    def acc(self, predictions, target):
        
        return np.where((predictions - target) == 0,1,0).sum() / len(predictions)
    
    def plot_decision_boundary(self, X, Y):
        
        b,w1,w2 = self.weights
        
        m = -(b / w2) / (b / w1)
        x = np.linspace(X[:,1:].min(),X[:,1:].max(), 100)
        y = m * x + (-b / w2)
        
        plt.plot(x, y, "-r")
        plt.scatter(X[:, 1], X[:, 2], c=Y)
        plt.xlabel("x", color="#1C2833")
        plt.ylabel("y", color="#1C2833")
        plt.grid()
        plt.show()
        
    def fit(self, X, y):
        
        
        # add bias
        X = np.concatenate((-np.ones((X.shape[0], 1)), X), axis=1)
        
        acc = []
        self.weights = np.random.randn(X.shape[1])

        for t in range(10):
            predictions = np.where(np.dot(self.weights, X.T) > 0,1,0)
            self.weights -= self.lr * np.dot((predictions - y), X)
            
            acc.append(self.acc(predictions, y))
            self.plot_decision_boundary(X,y)
               
model = perceptron()
model.fit(X, y)
```
